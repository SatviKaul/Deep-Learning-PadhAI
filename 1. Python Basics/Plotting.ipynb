```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.random((10, 1))
print(x)
```

```python
# line plotting
plt.plot(x)
plt.show()
```

```python
# star plotting
plt.plot(x, '*')
plt.show()
```

```python
# Star + line plotting
plt.plot(x, '*-')
plt.show()
```

```python
x = np.linspace(0, 10, 50)
y = np.power(x, 0.5) # power = 0.5 = square root
plt.plot(x, y, '*')
plt.show()
```

```python
import seaborn as sns
```
