```python
import numpy as np
import matplotlib.pyplot as plt
```

# Vector plotting

quiver() to plot a vector

```python
plt.quiver(0,0,3,4)
plt.show()
```

But the axis scaling isn't correct.
<br>To fix it we use further arguments.

```python
plt.quiver(0,0,3,4, scale_units="xy", angles="xy", scale=1)
plt.show()
```

But now it's going outside the window.
<br> So we extend the limits our dimensions.

```python
plt.quiver(0,0,3,4, scale_units="xy", angles="xy", scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
plt.quiver(0,0,3,4, scale_units="xy", angles="xy", scale=1, color="r")
plt.quiver(0,0,-3,4, scale_units="xy", angles="xy", scale=1, color="g")
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

### Write a function to print all the vectors in one go

```python
def plot_vectors(vecs):
    colors = ["r","g","b","y"]
    i = 0;
    for vec in vecs:
        plt.quiver(vec[0],vec[1],vec[2],vec[3], scale_units="xy", angles="xy", scale=1,color= colors[i%len(colors)])
        i=i+1
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()
```

```python
plot_vectors([(0,0,3,4), (0,0,-3,4), (0,0,2,-3), (0,0,-5,-3)])
```

```python
vecs = [(0,0,3,4), (0,0,-3,4), (0,0,2,-3), (0,0,-5,-3)]
```

```python
plot_vectors([vecs[0], vecs[2]])
```

# Vector  dot product

```python
vecs = [np.asarray([0,0,3,4]), np.asarray([0,0,-3,4]), np.asarray([0,0,-3,-2]), np.asarray([0,0,4,-1])]
```

```python
plot_vectors(vecs)
```

Taking 
<br>$ \vec{a} $ = vecs[0] = [0,0,3,4]
<br>$ \vec{b} $ = vecs[3] = [0,0,4,-1]

```python
a = np.asarray([3,4])
b = np.asarray([4,-1])
```

#### Dot product is given by-
$\vec{a}\cdot\vec{b} = |\vec{a}| |\vec{b}| \cos(\theta)$

```python
a_dot_b = np.dot(a,b)
```

```python
print(a_dot_b)
```

#### The projection of A onto B
$a_b = |\vec{a}|\cos(\theta) = |\vec{a}|\frac{\vec{a}\cdot \vec{b}}{|\vec{a}||\vec{b}|} = \frac{\vec{a} \cdot \vec{b}}{|\vec{b}|}$

```python
a_b = np.dot(a,b)/np.linalg.norm(b) # np.linalg.norm(b) gives mod(b)
```

```python
print(a_b)
```

#### To plot the projected vector
We have the magnitude now, we need direction to plot the projection vector. The direction of the projection vector will be in the direction of the $\vec{b}$.
<br>We use a unit vector along $\vec{b}$ i.e. $\hat{b}$.
<br>$\vec{a_b} = a_b\hat{b} = a_b\frac{\vec{b}}{|\vec{b}|}$

```python
vec_a_b = a_b/(np.linalg.norm(b))*b
print(vec_a_b)
```

```python
b_a = b_dot_a/np.linalg.norm(a)
```
